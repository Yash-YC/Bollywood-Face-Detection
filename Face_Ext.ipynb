```python
import cv2
import dlib
import os
import numpy as np
import glob
from PIL import Image
```

```python
actors = os.listdir("actors")
detector = dlib.get_frontal_face_detector()
new_path = "D:/python_ml/face/Actors_faces"
```

```python
filenames = []

for actor in actors:
    for file in os.listdir(os.path.join("actors", actor)):
        filenames.append(os.path.join("actors", actor, file))
filenames
```

```python
def save(img,name, bbox, width=180,height=227):
    try:
        x, y, w, h = bbox
        imgCrop = img[y:h, x: w]
        imgCrop = cv2.resize(imgCrop, (width, height))
        cv2.imwrite(name+".jpg", imgCrop)
    except Exception as e:
        print(str(e))
```

```python
def faces():
    for Counter,i in enumerate(filenames):
        try :       
            frame = cv2.imread(i)
            gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            faces = detector(gray)
            a = filenames[Counter].split("\\")[1]
        
            for face in faces:
                x1, y1 = face.left(), face.top()
                x2, y2 = face.right(), face.bottom()
                print(Counter) 
            # cv2.rectangle(frame,(x1,y1),(x2,y2),(220,255,220),1)
            # MyRec(frame, x1, y1, x2 - x1, y2 - y1, 10, (0,250,0), 3)
            # save(gray,new_path+str(counter),(x1-fit,y1-fit,x2+fit,y2+fit))
                save(frame, a + "." + str(Counter) , (x1,y1,x2,y2))
            frame = cv2.resize(frame,(800,800))
            cv2.imshow('',frame)
            cv2.waitKey(1)
            print("done saving")
        except Exception as e:
            print(str(e))
faces()    
```
